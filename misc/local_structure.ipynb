# Summary of "Dengue diversity across spatial and temporal scales: local structure and the impact of host population size"

### Authors: Henrik Salje et al. (Science, 2017)

This study explores dengue transmission dynamics in Bangkok, Thailand, using geolocated case data, serotype data, and phylogenetic analysis. The goal is to understand how the effective number of transmission chains scales with population density and distance, and how viral strains interact across spatial and temporal scales. The study uses phylogenetic trees and case data to evaluate dengue dispersal, providing insights into the spatial distribution of transmission chains and implications for vector control.

#### Key Findings:

- **Local Transmission Structure**: 60% of dengue cases within 200 meters in Bangkok belonged to the same transmission chain, compared to only 3% of cases separated by 15 km.
- **Population Size and Transmission Chains**: The effective number of transmission chains increased linearly with population size up to a certain threshold. For populations larger than 7,000 individuals/km, additional density did not lead to more chains, suggesting a saturation effect.
- **Regional and International Flow**: Within Thailand, viral lineages mixed extensively by the next dengue season, with limited flow between Thailand and neighboring countries, indicating largely endemic transmission.
